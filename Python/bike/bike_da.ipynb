```python
def ret_five(place_name):
    seocho = mb[mb['x_소재지'].str.contains(place_name)]
    r_seo = seocho.loc[:, ['a_위도','a_경도','a_대여소명']]
    rs = seocho.groupby(by= ['a_대여소명']).sum()
    s_bike = r_seo.drop_duplicates(['a_대여소명','a_위도','a_경도'], keep = 'first')
    seocho_join = pd.merge(rs,s_bike, left_on='a_대여소명', right_on ='a_대여소명', how = 'right')
    seocho_top5 = seocho_join.sort_values(by = "x_반납건수", ascending=False)
    return seocho_top5.head(5)
```

```python
import folium
import pandas as pd
geo_json = 'https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
rent = pd.read_csv('result.csv')
rent

def ret_five(place_name):
    seocho = mb[mb['x_소재지'].str.contains(place_name)]
    r_seo = seocho.loc[:, ['a_위도','a_경도','a_대여소명']]
    rs = seocho.groupby(by= ['a_대여소명']).sum()
    s_bike = r_seo.drop_duplicates(['a_대여소명','a_위도','a_경도'], keep = 'first')
    seocho_join = pd.merge(rs,s_bike, left_on='a_대여소명', right_on ='a_대여소명', how = 'right')
    seocho_top5 = seocho_join.sort_values(by = "x_반납건수", ascending=False)
    return seocho_top5.head(5)
```

```python
m = folium.Map()
folium.Choropleth(
    geo_data= geo_json,
    name ='choropleth',
    data = rent,
    columns = ['소재지','반납건수'],
    key_on = 'feature.properties.name',
    fill_color='PuRd',
    fill_opacity=0.7,
    line_opacity=0.1,
).add_to(m)
```

```python
def add_five(place_name):
    test = ret_five(place_name)

    xs = list(test["a_위도"])
    ys = list(test["a_경도"])
    titles = list(test["a_대여소명"])
    cnts = list(test["x_반납건수"])

    for i in range(5):
        
        # if i == 0:

        folium.Map(location=[xs[i],	ys[i]], zoom_start=16, titles = 'stemenwatercolor')#위도,경도
        toolitp = titles[i]
        folium.Marker([xs[i],	ys[i]], popup='<i>{}건<i>'.format(cnts[i]), tooltip=toolitp).add_to(m)
```

```python
place_set = list(set(list(mb["x_소재지"])))
place_set
```

```python
for place in place_set:
    try:
        add_five(place)
    except:
        pass
```
