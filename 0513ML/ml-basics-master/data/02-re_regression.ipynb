```python
import pandas as pd
import matplotlib.pyplot as plt

#플롯이 쥬피터에 inline으로 표시됨
%matplotlib inline

#레이블 열 가져오기
label = bike_data['rentals']

#2개의 서브플롯(2행,1열) 그림 생성
fig, ax =  plt.subplots(2, 1, figsize = (9,12))

#히스토그램 플로팅 ##bins-> 몇개의 영역으로 쪼갤래? 100개
#set_ylabel-> y 축 설정하기 ##ax ->좌표평면 결정
ax[0].hist(label, bins = 100)
ax[0].set_ylabel('Frequency')

#axvline->지정 위치에 수직 수평선 그려주기 #linewidth->선의 넓이 조정
#ax[0]이라는 액자에 그림을 그릴래
ax[0].axvline(label.mean(), color = 'magenta', linestyle='dashed', linewidth = 2)
ax[0].axvline(label.median(), color = 'cyan', linestyle='dashed', linewidth = 2)


#결측치가 동그라미 그림으로 나타남inewidth = 2)
ax[0].axvline(label.median(), color = 'cyan', linestyle='dashed', linewidth = 2)

# vert=False 가로축그리기
#동그라미 말고 그냥 길쭉한 애는 뭐하는 애야? 평균을 나타냄동그라미 말고 그냥 길쭉한 애는 뭐하는 애야? 평균을 나타냄
ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# fig.suptitle->그림에 제목 추가
fig.suptitle('Remtal Distribution')


ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

ax[1].boxplot(label, vert=False)
ax[1].set_xlabel('Rentals')

# fig.suptitle->그림에 제목 추가
fig.suptitle('Remtal Distribution')

#그림 보여줘
fig.show()
```

#### 플롯에 따르면 일일 임대 횟수가 0-3400약간 넘음. 그러나 일일 대여 평균은 0-2200건 사이임, 동그라미는 이상치를 뜻함:비정상적으로 높거나 낮은값

```python
# Plot a histogram for each numeric feature
for col in numeric_features:
    fig = plt.figure(figsize =(9,6)) #figure는 흰네모 자체를 만드는 객체
    ax = fig.gca()#gca()->gca()가 있는 현재의 Axes 객체를 구해준다.
    feature = bike_data[col]
    #ax(axes)->figure 내에서 축을 가지는 하나의 좌표평면과 같은 개념
    feature.hist(bins=100, ax = ax)
    ax.axvline(feature.mean(), color='magenta', linestyle = 'dashed', linewidth=2)
    ax.axvline(feature.median(), color='cyan', linestyle = 'dashed', linewidth=2)
    ax.set_title(col)#set->col명을 설정해줘
plt.show()
```

####  위 그림은for문을 활용해서 컬럼별로 히스토그램 만들어보기
#### temp와 atemp는 분포의모양이 좋지 않아서 히스토그램 사용X
#### hum 과 windspeed는 그나마 종모양에 가까움

```python
import numpy as numpy

# plot a bar plot for each categorical feature count
##categorical feature : 범주형 변수mport numpy as numpy

categorical_features=['season','mnth','holiday','weekday','workingday','weathersit', 'day']

for col in categorical_features:
    ##bike_data[col]->pd.Series ##value_counts()->유일한 값별 수세기
    ##sort_index() -> 데이터 큰값부터 정렬됨   ##sort_index() -> 데이터 큰값부터 정렬됨   
    counts = bike_data[col].value_counts().sort_index()
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()  #ax(axes)->figure 내에서 축을 가지는 하나의 좌표평면과 같은 개념
    counts.plot.bar(ax = ax, color='steelblue')# ->count변수를 플롯 바차드로 그려줘
    ax.set_title(col + 'counts')#-> 제목에 컬럼명이랑 같이 counts붙여서 출력해줘
    ax.set_xlabel(col) #->x축에는 col명을
    ax.set_ylabel("Frequency")# -> y축에는 빈도를 넣어주도록
plt.show()
```

```python
for col in numeric_features:
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    feature = bike_data[col]
    label = bike_data['rentals']
    correlation = feature.corr(label)
    plt.scatter(x=feature, y=label)
    plt.xlabel(col)
    plt.ylabel('Bike Rentals')
    ax.set_title('rentals vs ' + col + '-correlation ' + str(correlation))
    plt.show()
```

#### 결과가 결정적이지는 않으나 온도가 높으면 자전거 대여수도 많아짐
#### 풍속이 높은 날에는 자전거 대여가 거의 없음

```python
# plot a boxplot for the label by each categorical feature
#각 범주형 특징을 기준으로 레이블에 대한 상자 그림 그리기
#label -> 예측을 하고자 하는 대상 항목(=종속변수y)
for col in categorical_features:
    fig = plt.figure(figsize=(9,6))
    ax = fig.gca()
    bike_data.boxplot(column = 'rentals', by = col, ax = ax)
    ax.set_title('Label by' + col)
    ax.set_ylabel("Bike Rentals")
plt.show()
```
